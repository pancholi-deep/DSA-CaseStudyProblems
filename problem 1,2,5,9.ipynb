```python
from statistics import mean
from statistics import variance
from math import sqrt

class Tstat:
    def __init__(self,l1,l2):
        self.l1 = l1
        self.l2 = l2
        
    def get_mean(self,l):
        return mean(l)
    
    def get_variance(self,l):
        return variance(l)
  
    def get_tstat(self,m1,m2,v1,v2,n):
        denom = sqrt((v1+v2)/n)
        numer = m1 - m2
        return numer/denom
```

```python
N = int(input())
n = N
a = []
b = []
while n>0:
    x = list(map(int, input().split()))
    a.append(x[0])
    b.append(x[1])
    n -= 1
    
T = Tstat(a,b)
m1,v1 = T.get_mean(a), T.get_variance(a)
m2,v2 = T.get_mean(b), T.get_variance(b)
print(f"first numbers in pairs: Mean = {m1}, Variance = {v1}")
print(f"second numbers in pairs: Mean = {m2}, Variance = {v2}")
print(f"t-statistic between two lists = {T.get_tstat(m1,m2,v1,v2,N)}")

"""
 i 	xi 	yi 
1	 10.0 	14.0 
2 	12.0 	14.0 
3 	9.0 	14.0 
4 	9.0 	18.0

"""
```

```python
class HVACLoad:
    
    def __init__(self, roof_area, b_perimeter, b_height, w_size, west_w_size, roof_u, wall_u, window_u, 
                 winter_o_temp, summer_o_temp, summer_o_humi, infil, people, elect_use):
        self.roof_area = roof_area
        self.b_perimeter = b_perimeter
        self.b_height = b_height
        self.w_size = w_size
        self.west_w_size = west_w_size
        self.roof_u = roof_u
        self.wall_u = wall_u
        self.window_u = window_u
        self.winter_o_temp = winter_o_temp
        self.summer_o_temp = summer_o_temp
        self.summer_o_humi = summer_o_humi
        self.infil = infil
        self.people = people
        self.elect_use = elect_use
        
    def findHeatingLoad(self):
        winter_i_temp = 72
        
        winter_temp_diff = winter_i_temp - self.winter_o_temp
        net_wall_area = (self.b_perimeter * self.b_height) - self.w_size

        roof_load = self.roof_area * self.roof_u * winter_temp_diff
        window_load = self.w_size * self.window_u * winter_temp_diff
        wall_load = net_wall_area * self.wall_u * winter_temp_diff
        infiltration_load = self.infil * 1.08 * winter_temp_diff
        
        heating_load = 1.3*(roof_load + window_load + wall_load + infiltration_load)
        return heating_load
    
    def findCoolingLoad(self): 
        summer_i_humi = 0.01
        
        summer_humidity_diff = self.summer_o_humi - summer_i_humi
        
        infil_humidity_load = self.infil * 4675 * summer_humidity_diff
        people_humidity_load = 200 * self.people
        
        cooling_load = 1.1*(self.findCoolingSensibleLoad() + infil_humidity_load + people_humidity_load)
        
        return cooling_load

    def findCoolingSensibleLoad(self):
        summer_i_temp = 76
        sunshine_west_w = 193

        net_wall_area = (self.b_perimeter * self.b_height) - self.w_size
        summer_temp_diff =  self.summer_o_temp - summer_i_temp
        
        roof_load = self.roof_area * self.roof_u * summer_temp_diff
        window_load = self.w_size * self.window_u * summer_temp_diff
        wall_load = net_wall_area * self.wall_u * summer_temp_diff
        infiltration_load = self.infil * 1.08 * summer_temp_diff
        solar_load = sunshine_west_w * self.west_w_size
        e_load = 3.416 * self.elect_use
        people_temp_load = 250 * self.people
        
        sensible_load = roof_load + window_load + wall_load + infiltration_load + solar_load + e_load + people_temp_load
        return sensible_load
```

```python
class Queue:
    def __init__(self):
        self.items = []
 
    def enqueue(self, data):
        self.items.append(data)
        
    def dequeue(self):
        return self.items.pop(0)
    
    def first(self):
        return self.items[0]
    
    def size(self):
        return len(self.items)

def lastCustomer(m, n):
    q = Queue()
    customers = m+1
    
    for i in range(1, customers):
        q.enqueue(i)
    print(q.items)

    for i in range(customers,0,-1):
        for _ in range(n,0,-1):
            if q.size() == 1:
                return q.first()
            print(f"Customer {q.first()} sent back.")
            temp = q.dequeue()
            q.enqueue(temp)
            print(q.items)
        
        print(f"Served Customer {q.first()}.")
        q.dequeue()
```

```python
m = int(input("Enter the number of customers: "))
n = int(input("Enter the number of rotations of queue: "))
print(f"The last customer to be served: {lastCustomer(m,n)}.")
```
